**Deep learning** is a subset of machine learning that uses neural networks with many layers inspired by the **human brain** _( which is also the reason it's called **"deep"**)_ to model **complex** & **nonlinear** patterns in data. It has been particularly successful in areas such as **image** and **speech recognition**, **natural language processing**, and more. even though Deep learning is a subset of machine learning it still has its major diffrences with it that you will see later on this file.

**Deep learning** and **machine learning** are both pivotal components of **artificial intelligence**, with **deep learning** being a specialized subset of **machine learning**. While they share the common **goal** of enabling machines to **learn** from **data**, they each bring unique methodologies and capabilities to the table. Understanding the **nuances** between these two approaches is essential for appreciating their strengths and applications. In the following sections, we will explore how **deep learning** and **machine learning** differ, and how these differences impact their use in various fields.

**Machine learning** can work effectively with **smaller datasets**, making it suitable for a wide range of applications. In contrast, **deep learning** requires **large amounts of data** to perform well, as it relies on **extensive data** to learn intricate patterns and features.

**Machine learning** generally requires **less** computational power and can run on **standard CPUs**. **Deep learning**, on the other hand, demands **significant** computational resources, often necessitating the use of **GPUs** or specialized hardware like **TPUs** to handle the **complex computations** involved.

Artificial Neural Networks **(ANNs)** are the fundamental building blocks of **deep learning**, inspired by the structure of the **human brain**. They consist of interconnected **nodes**, or **neurons**, organized into **layers**. Each neuron processes **input data** and passes the **output** to the next **layer**, allowing the network to **learn** from data, **identify patterns**, and make **predictions** or **decisions**.

<img src="Img/ann-min.jpg" alt="Deep Learning in Python" width="750" height="400">

**ANNs** typically include an **input layer**, one or more **hidden layers**, and an **output layer**. The **input layer** receives the raw data, which is then **passed** through the **hidden layers** where the **actual processing** and **learning occur**. Each **hidden layer** consists of multiple **neurons** that apply transformations to the data. Finally, the **output layer** produces the result, such as a classification or prediction. This layered structure enables **ANNs** to model **complex relationships** and perform a wide range of tasks, from simple regression and classification to more advanced applications like image and speech recognition. 

<img src="Img/Types-of-Neural-Networks-min.jpg" alt="Deep Learning in Python" width="500" height="47">

<img src="Img/Feedforward-Neural-Networks-(FNN)-min.jpg" alt="Deep Learning in Python" width="380" height="47">

The **neural network** you saw **in the previous section** is a **Feedforward Neural Network**, Feedforward Neural Networks **(FNN)** are the **simplest** type of artificial neural network. In an **FNN**, data moves in **one direction from** the **input layer**, through the **hidden layers**, and finally to the **output layer**. There are **no cycles** or **loops** in the network. Each **neuron** in one layer is connected to **every neuron** in the **next layer**, allowing the network to learn **complex patterns** by adjusting the **weights** of these connections **during training**. **FNNs** are commonly used for tasks like image classification and regression analysis, where the relationship between input and output is **straightforward**.

<img src="Img/Recurrent-Neural-Networks-(RNN)-min.jpg" alt="Deep Learning in Python" width="380" height="47">

Recurrent Neural Networks **(RNN)** are designed to handle **sequential data**, where the **order** of the data points is important.

<img src="Img/RNN-FNN.jpg" alt="Deep Learning in Python" width="750" height="400">

Unlike **feedforward** networks, **RNNs** have connections that form **directed cycles**, allowing information to persist across time steps. This makes **RNNs** well-suited for tasks like **time series prediction**, **natural language processing**, and **speech recognition**.

<img src="Img/Convolutional-Neural-Networks-(CNN)-min.jpg" alt="Deep Learning in Python" width="380" height="47">

Convolutional Neural Networks **(CNN)** are specialized **neural networks** designed for processing **structured grid data**, such as **images**. **CNNs** use a series of **convolutional** layers that apply **filters** to the **input data**, capturing spatial hierarchies and patterns. These layers are followed by **pooling layers** that **reduce the dimensionality** of the data, making the network **more efficient**. **CNNs** are particularly effective for **image-related tasks**, such as object detection, facial recognition, and medical image analysis, due to their ability to automatically learn and extract relevant features from raw pixel data.

**Activation functions** introduce **non-linearity** into the **neural network**, enabling it to learn and represent complex patterns in the data. They determine whether a **neuron** should be activated by calculating the **weighted sum** of inputs and applying a **mathematical function**. Common activation functions include **ReLU**, **Sigmoid**, and **Tanh**. We will explore **activation functions** in detail and how they work.

<img src="Img/sigmoid-min.jpg" alt="Deep Learning in Python" width="300" height="47">

The **sigmoid** function takes any **real-valued** number and **squashes** it into a range between **0** and **1**. This makes it useful for transforming the output of a neuron into a probability-like value the formula is written below and the shape is showcased. <span style="font-family: Arial; font-size: 14px;">
$$\sigma(x) = \frac{1}{1 + e^{-x}}$$
</span>

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")


def sigmoid(x):
    return 1 / (1 + np.exp(-x))


x_vals = np.linspace(-10, 10, 600)
plt.figure(figsize=(32, 8))
plt.subplot(2, 4, 1)
sns.lineplot(x=x_vals, y=sigmoid(x_vals), color='gold')
plt.show()
```

<img src="Img/Hyperbolic-Tangent-min.jpg" alt="Deep Learning in Python" width="300" height="47">

It **maps** input values to a range between **-1** and **1**, making it **zero-centered**. the formula is written below and the shape is showcased.
<span style="font-family: Arial; font-size: 14px;">
$$\tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}$$
</span>

```python
def tanh(x):
    return np.tanh(x)
    
plt.figure(figsize=(32, 8))
plt.subplot(2, 4, 2)
sns.lineplot(x=x_vals, y=tanh(x_vals), color='lime')
plt.show()
```

<img src="Img/RELU-min.jpg" alt="Deep Learning in Python" width="300" height="47">

The **ReLU** function outputs the **input** directly if it is **positive**; otherwise, it outputs **zero**. the formula is written below and the shape is showcased. <span style="font-family: Arial; font-size: 14px;">
$$\text{ReLU}(x) = \max(0, x)$$
</span>

```python
def relu(x):
    return np.maximum(0, x)
    
plt.figure(figsize=(32, 8))
plt.subplot(2, 4, 2)
sns.lineplot(x=x_vals, y=relu(x_vals), color='cyan')
plt.show()
```

<img src="Img/Leaky-ReLU-min.jpg" alt="Deep Learning in Python" width="300" height="47">

 **Leaky ReLU** is a **variation** of **ReLU** that allows a small, **non-zero** gradient when the input is **negative**. the formula is written below and the shape is showcased. <span style="font-family: Arial; font-size: 14px;">
$$\text{Leaky ReLU}(x) = \begin{cases} 
      x & \text{if } x > 0 \\
      \alpha x & \text{if } x \leq 0 
   \end{cases}$$
</span>

```python
def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, x * alpha)
    
plt.figure(figsize=(32, 8))
plt.subplot(2, 4, 2)
sns.lineplot(x=x_vals, y=leaky_relu(x_vals), color='green')
plt.show()
```

<img src="Img/Softmax Function-min.jpg" alt="Deep Learning in Python" width="300" height="47">

The **softmax** function is used in the output layer of neural networks for **multi-class** **classification**. It converts **logits** into **probabilities**. the formula is written below and the shape is showcased. <span style="font-family: Arial; font-size: 14px;">
$$\text{Softmax}(x_i) = \frac{e^{x_i}}{\sum_{j} e^{x_j}}$$
</span>

```python
def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum(axis=0)
    
plt.figure(figsize=(32, 8))
plt.subplot(2, 4, 2)
sns.lineplot(x=x_vals, y=softmax(x_vals), color='yellow')
plt.show()
```

<img src="Img/ELU-min.jpg" alt="Deep Learning in Python" width="300" height="47">

**ELU** is similar to **ReLU** but tends to converge **faster** and produce more **accurate results**. the formula is written below and the shape is showcased. <span style="font-family: Arial; font-size: 14px;">
$$\text{ELU}(x) = \begin{cases} 
      x & \text{if } x > 0 \\
      \alpha (e^x - 1) & \text{if } x \leq 0 
   \end{cases}$$
</span>

```python
def elu(x, alpha=1.0):
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))
    
plt.figure(figsize=(32, 8))
plt.subplot(2, 4, 2)
sns.lineplot(x=x_vals, y=elu(x_vals), color='red')
plt.show()
```

<img src="Img/swish-min.jpg" alt="Deep Learning in Python" width="300" height="47">

**Swish** is a **smooth**, **non-monotonic** activation function. the formula is written below and the shape is showcased. <span style="font-family: Arial; font-size: 14px;">
$$\text{Swish}(x) = x \cdot \sigma(x)$$
</span>

```python
def swish(x):
    return x * sigmoid(x)
    
plt.figure(figsize=(32, 8))
plt.subplot(2, 4, 2)
sns.lineplot(x=x_vals, y=swish(x_vals), color='purple')
plt.show()
```

<img src="Img/mish-min.jpg" alt="Deep Learning in Python" width="300" height="47">

**Mish** is a newer activation that combines properties of both **Swish** and **Tanh**. the formula is written below and the shape is showcased.<span style="font-family: Arial; font-size: 14px;">
$$\text{Mish}(x) = x \cdot \tanh(\ln(1 + e^x))$$
</span>

```python
def mish(x):
    return x * np.tanh(np.log1p(np.exp(x)))
    
plt.figure(figsize=(32, 8))
plt.subplot(2, 4, 2)
sns.lineplot(x=x_vals, y=swish(x_vals), color='darkblue')
plt.show()
```

**Loss functions**, also known as **cost functions**, measure how well a neural network’s predictions match the actual outcomes. They play a critical role in **training** by guiding the **optimization** process. **Common loss functions** include **Mean Squared Error (MSE)** for regression tasks, which calculates the **average squared** difference between predicted and actual values, and **Cross-Entropy Loss** for **classification tasks**, which measures the difference between the predicted probability distribution and the true distribution. The choice of loss function depends on the type of problem being solved and directly impacts the performance of the model.

<img src="Img/opti-min.jpg" alt="Deep Learning in Python" width="400" height="47">

**Optimization algorithms** are used to **minimize** the **loss function** by adjusting the **weights** of the neural network. **Gradient Descent** is a fundamental optimization algorithm that iteratively updates the **weights** in the direction of the **negative gradient** of the **loss function**. Variants like **Stochastic Gradient Descent (SGD)** update **weights** using a **single data** point at a time, making it **faster** for **large datasets**. Advanced algorithms like **Adam** (Adaptive Moment Estimation) combine the advantages of both **SGD** and **RMSprop**, adjusting the learning rate based on the first and second moments of the gradients, leading to **faster** and more **reliable** convergence.
